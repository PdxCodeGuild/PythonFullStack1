# A Better way to opening files: Context Manager!


```python
"""
Most commonly, files are opened inside of a `with` block.
This will add some additional behavior to solve the problem we encountered above.
"""

# readline - One line at a time as an iterator.
# readlines - Returns a list split by newline.
# read returns one string of the file contents.

with open('../../labs/projects/ari/books/odyssey.txt', 'r') as file:
    text = file.read()
    print(text)
        
        
```
