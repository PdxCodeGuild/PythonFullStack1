```python
class Wall:
    def __init__(self, height, width):
        self.width = width
        self.height = height
        
        self.is_painted = False
        self.area = width * height
        
    def __repr__(self):
        return "a {} by {} wall.".format(self.width, self.height)
    
    def paint(self, cost, coverage=400):
        result = (self.area / coverage) * cost
        print("OK we painted it")
        self.is_painted = True
        
        return result
       
    
```

```python
first_wall = Wall(10, 10)
second_wall = Wall(12, 8)
```

```python
first_wall.paint(4.5)
```

```python
first_wall.is_painted
```

```python
walls = [first_wall, second_wall]
```

```python
walls
```

```python
[wall.paint(3.0) for wall in walls]
```
